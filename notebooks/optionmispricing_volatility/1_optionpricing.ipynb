# Option Prices

```python
df_options_btc = pd.read_parquet('/Users/ryant/Documents/Github/digital-asset-options-pricing/src/data/deribit/data/deribit_btc_snapshot_011124_301124.parquet')
df_options_btc.head(5)
```

## Data Combined

```python
df_options = df_options_btc[['date', 'instrument_name', 'expiration_date', 'strike', 'claim_type', 'underlyer_spot', 'best_bid_price', 'best_ask_price', ]].copy()

df_options['datetime'] = pd.to_datetime(df_options['date'], unit='ms', utc=True)
# df_options['date'] = df_options.datetime.dt.date
df_options['hour'] = df_options.datetime.dt.hour
df_options['minute'] = df_options.datetime.dt.hour

df_options_daily = df_options[(df_options['hour'] == 0) & (df_options['minute'] == 0)].reset_index(drop=True)
df_options_daily['expiration'] = pd.to_datetime(df_options_daily['expiration_date']).dt.strftime('%d-%m-%Y')
df_options_daily['calculation_date'] = pd.to_datetime(df_options_daily['date'], unit='ms', utc=True).dt.strftime('%d-%m-%Y')
df_options_daily.loc[:,'dividend_rate'] = 0
df_options_daily.loc[:,'rf_rate'] = 0.0
df_options_daily['days_to_maturity'] = (pd.to_datetime(df_options_daily['expiration_date']) - df_options_daily['datetime']).dt.days
df_options_daily = df_options_daily.merge(df_binance_btc[['timestamp', 'ann_parkinson_vol']], left_on='date', right_on='timestamp').drop(columns=['timestamp']).rename(columns={'ann_parkinson_vol': 'sigma'})
df_options_daily = df_options_daily.rename(columns = {'underlyer_spot' : 'underlying'})
df_options_daily[['strike', 'underlying']] = df_options_daily[['strike', 'underlying']].astype(float)
df_options_daily['lowest_bid'] = df_options_daily['best_bid_price'] * df_options_daily['underlying']
df_options_daily['highest_ask'] = df_options_daily['best_ask_price'] * df_options_daily['underlying']
df_options_daily['mid_price'] = (df_options_daily['lowest_bid'] + df_options_daily['highest_ask']) / 2
df_options_daily
```

## Pricing (1 Instrument)

```python
instrument = df_options_daily.iloc[2]['instrument_name']
df_options_single_timestamp = df_options_daily[df_options_daily['instrument_name'] == instrument].reset_index(drop=True)
df_options_single_timestamp.head(5)
```

```python

pricer = AnalyticEuropeanPricer(option_type=ql.Option.Call if instrument[-1] == 'C' else ql.Option.Put)
# analyticeuropean_pricer = AnalyticEuropeanPricer(option_type=ql.Option.Call if instrument[-1] == 'C' else ql.Option.Put)

df_options_single_timestamp['premium'] = df_options_single_timestamp.apply(lambda x: pricer.price(
    calculation_date = ql.Date(x['datetime'].day, x['datetime'].month, x['datetime'].year),
    underlying = x['underlying'],
    strike = x['strike'],
    sigma = x['sigma'],
    rf_rate = x['rf_rate'],
    days_to_maturity = x['days_to_maturity'],
    div = x['dividend_rate']
), axis=1)
df_options_single_timestamp
```

## Price All Instruments

```python

analyticeuropean_call_pricer = AnalyticEuropeanPricer(option_type=ql.Option.Call)
analyticeuropean_put_pricer = AnalyticEuropeanPricer(option_type=ql.Option.Put)

steps = 100

binomialeuropean_call_pricer = BinomialEuropeanPricer(steps = steps, option_type=ql.Option.Call)
binomialeuropean_put_pricer = BinomialEuropeanPricer(steps = steps, option_type=ql.Option.Put)

num_paths = 500

mceuropean_call_pricer = MCEuropeanPricer(steps = steps, num_paths=num_paths, option_type=ql.Option.Call)
mceuropean_put_pricer = MCEuropeanPricer(steps = steps, num_paths=num_paths, option_type=ql.Option.Put)
```

```python
df_options_daily_calls = df_options_daily[df_options_daily['claim_type'] == 'call'].reset_index(drop=True)
df_options_daily_puts = df_options_daily[df_options_daily['claim_type'] == 'put'].reset_index(drop=True)

call_pricers = {
    'analyticeuropean_call_premium': analyticeuropean_call_pricer,
    'binomialeuropean_call_premium': binomialeuropean_call_pricer,
    'mceuropean_call_premium': mceuropean_call_pricer
}
put_pricers = {
    'analyticeuropean_put_premium': analyticeuropean_put_pricer,
    'binomialeuropean_put_premium': binomialeuropean_put_pricer,
    'mceuropean_put_premium': mceuropean_put_pricer
}

for n, p in call_pricers.items():
    df_options_daily_calls[n] = df_options_daily_calls.apply(lambda x: p.price(
        calculation_date = ql.Date(x['datetime'].day, x['datetime'].month, x['datetime'].year),
        underlying = x['underlying'],
        strike = x['strike'],
        sigma = x['sigma'],
        rf_rate = x['rf_rate'],
        days_to_maturity = x['days_to_maturity'],
        div = x['dividend_rate']
    ), axis=1)

for n, p in put_pricers.items():
    df_options_daily_puts[n] = df_options_daily_puts.apply(lambda x: p.price(
        calculation_date = ql.Date(x['datetime'].day, x['datetime'].month, x['datetime'].year),
        underlying = x['underlying'],
        strike = x['strike'],
        sigma = x['sigma'],
        rf_rate = x['rf_rate'],
        days_to_maturity = x['days_to_maturity'],
        div = x['dividend_rate']
    ), axis=1)
```
